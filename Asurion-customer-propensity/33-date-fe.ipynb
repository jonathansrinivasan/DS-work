### Date Transformation for all encoded data

#### There was also transformation for some variables that needed their data types refined

```python
def date_transform(data):
    df = pd.read_csv(data)
    df["clndr_dt_cst"] = pd.to_datetime(df["clndr_dt_cst"], format = "%Y-%m-%d")
    df["start_dt"] = pd.to_datetime(df["start_dt"], format = "%Y-%m-%d")
    df["zipcode"] = df["zipcode"].astype('Int64')
    df = df.drop(["lat_long_source","accepted_flg_1"], axis = 1)
    df["day_of_week"] = df['clndr_dt_cst'].dt.day_name()
    df["time_of_day"] = df['start_dt'].dt.hour
    df.to_csv(data, index = False)
```
