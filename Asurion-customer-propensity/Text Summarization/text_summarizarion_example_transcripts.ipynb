```python
sample_text = df.encrypted_collated_transcription[0]#take the first transcript to test out
summaries = {} #define this to store all the output summaries
```

```python
#PEGASUS
def pegasus_summarization(text):
  pipe_pega = pipeline("summarization", model="google/pegasus-cnn_dailymail")
  pipe_out_pega = pipe_pega(text)
  summaries["pegasus"] = pipe_out_pega[0]["summary_text"].replace(" .<n>", ".\n")
```

```python
#BART
def Bart_summarization(text):
  pipe_bart = pipeline("summarization", model="facebook/bart-large-cnn")
  pipe_out_bart = pipe_bart(text)
  summaries["bart"] = "\n".join(sent_tokenize(pipe_out_bart[0]["summary_text"]))
```

```python
pegasus_summarization(sample_text)
```

```python
Bart_summarization(sample_text)
```
