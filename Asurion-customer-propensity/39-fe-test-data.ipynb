### Regular FE

```python
df_test = pd.read_csv("data/tabular_july_encoded.csv").clean_names()
df_test.info()
```

```python
columns_to_int = ['seven_day_churn', 'thirty_day_churn','ninety_day_churn', 'lines', 'account_tenure',
                  'num_decline', 'claritas_hh_count']
```

```python
for i in range(len(columns_to_int)):
    df_test[columns_to_int[i]] = df_test[columns_to_int[i]].astype('Int64')
```

```python
df_test.info()
```

### Income FE

```python
df_test['hh_income_code'] = df_test['hh_income'].replace(['1 Wealthy', '2 Upscale', '3 Upper Midscale', '4 Midscale', '5 Lower Midscale', '6 Downscale', '7 Low Income', 'Unknown'], [1, 2, 3, 4, 5, 6, 7, 8])
```

### Date FE

```python
df_test["clndr_dt_cst"] = pd.to_datetime(df_test["clndr_dt_cst"], format = "%Y-%m-%d")
df_test["start_dt"] = pd.to_datetime(df_test["start_dt"], format = "%Y-%m-%d")
df_test = df_test.drop("accepted_flg_1", axis = 1)
df_test["day_of_week"] = df_test['clndr_dt_cst'].dt.day_name()
df_test["time_of_day"] = df_test['start_dt'].dt.hour
```

```python
df_test = df_test.drop("unnamed_0", axis = 1)
```

```python
df_test["month"] = "july"
```

```python
df_test.info()
```
