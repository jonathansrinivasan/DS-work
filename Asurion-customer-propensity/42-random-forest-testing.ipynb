```python
from sklearn.datasets import make_classification
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.ensemble import RandomForestClassifier
```

```python
Y = df[["quality_sale"]].values 
```

```python
print(Y.shape)
```

```python
X = df.drop(labels = ["quality_sale"], axis=1) 
```

```python
print(X.shape)
```

```python
#Splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.25, random_state=20)
```

```python
print("X_train shape: ", X_train.shape)
print("y_train shape: ", y_train.shape)
print("X_test shape: ", X_test.shape)
print("y_test shape: ", y_test.shape)
```

```python
# Instantiate model with 10 decision trees
model = RandomForestClassifier(n_estimators = 10, random_state = 1)
# Train the model on training data
model.fit(X_train, y_train)
```

```python
prediction_test = model.predict(X_test)
```

```python
from sklearn.metrics import precision_score, f1_score, recall_score, confusion_matrix, accuracy_score, roc_auc_score
#Print the prediction accuracy
print ("Accuracy = ", accuracy_score(y_test, prediction_test))
print ("Precision = ", precision_score(y_test, prediction_test))
print ("Recall = ", recall_score(y_test, prediction_test))
print ("Confusion_matrix ", confusion_matrix(y_test, prediction_test))
```

From the result of a baseline Random Forest classifier, we can see that the results are significantly bad with the low recall score. Also, the false negative value in the confusion matrix is also something that we need to think about, the model is taking a lot of the postive values(quality_sale = 1) to be predicted as negative values(quality_sale = 0). This is probably because that the data is significantly imbalanced. So we need to apply some adjustments to deal the problem that the data is imbalanced.

```python
model_with_class_weight = RandomForestClassifier(n_estimators=10, class_weight='balanced_subsample',random_state = 1)
```

```python
model_with_class_weight.fit(X_train, y_train)
```

```python
prediction_test_with_weight = model_with_class_weight.predict(X_test)
```

```python
print ("Accuracy = ", accuracy_score(y_test, prediction_test_with_weight))
print ("Precision = ", precision_score(y_test, prediction_test_with_weight))
print ("Recall = ", recall_score(y_test, prediction_test_with_weight))
print ("Confusion_matrix ", confusion_matrix(y_test, prediction_test_with_weight))
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
#Hyperparameter tuning
n_estimators = [10,20,30,40,50]
max_depth = [1,2,3,4,5]
min_samples_split = [2, 5, 10, 15, 100]
min_samples_leaf = [1, 2, 5, 10] 

hyperF = dict(n_estimators = n_estimators, max_depth = max_depth,  
              min_samples_split = min_samples_split, 
             min_samples_leaf = min_samples_leaf)

gridF = GridSearchCV(model_with_class_weight, hyperF, cv = 3, verbose = 1, 
                      n_jobs = -1, scoring = 'recall')
bestF = gridF.fit(X_train, y_train)
```

```python
print(bestF.best_params_)
```
