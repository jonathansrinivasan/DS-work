```python
flgs = ['quality_sale', 'accepted_flg', 'seven_day_churn',
       'thirty_day_churn', 'ninety_day_churn', 'exp']
dates = ['clndr_dt_cst', 'start_dt']
num_list = ['lines', 'previous_tech_support_call', 'account_tenure', 'past_yr_claim_count', 'num_decline', 
            'claritas_hh_count', 'median_income', 'total_population', 'perc_female', 'perc_under18',
            'perc_over65', 'perc_white', 'perc_black', 'population_density', 'latitude', 'longitude',
            'zipcode_accepted_prop', 'zipcode_rejected_offers', 'zipcode_accepted_offers', 'zipcode_total_offers']
cat_list = ['product_type', 'pmake', 'technodoption_class', 'hh_income', 'age_class', 'hh_education', 
            'hh_composition', 'hh_tenure', 'urbanicity', 'month', 'day_of_week', 'time_of_day']
others = ['zipcode', 'county', 'state', 'state_code', 'hh_income_code']
```

```python
for i in num_list:
    print(i+' missing rate:',total_encoded_tabular[i].isnull().sum()/len(total_encoded_tabular[i]))
```

```python
for i in num_list:
    total_encoded_tabular[i] = total_encoded_tabular[i].fillna(0)
```

```python
total_encoded_tabular.shape[1] == len(flgs) + len(dates) + len(num_list) + len(cat_list) + len(others)
```

### Categorical Variables

```python
for col in cat_list:
    print(col+':', total_encoded_tabular[col].nunique(), total_encoded_tabular[col].unique(), '\n')
```

```python
# one hot encoding
total_encoded_tabular = pd.get_dummies(total_encoded_tabular, columns=cat_list, prefix=cat_list)

def rename_col(col:str):
    col = col.lower() # convert names to all lowercase
#     col = re.sub('[{}]'.format(string.punctuation), " ", col) # remove punctuations
    col = ' '.join(col.split()) # fix whitespaces
    col = col.replace(' ', '_') # replace spaces with underscores
    col = col.replace('-', '_')
    return col
for col in total_encoded_tabular.columns:
    total_encoded_tabular.rename(columns={col:rename_col(col)}, inplace=True)
```

### Numerical Variables

```python
def mynorm(df, num_list, norm_method):
    # norm_method = True zscore, False min-max
    for col in num_list:
        df[col+'_normalized'] = df[col].transform(lambda x: (x - x.mean()) / x.std()) if norm_method == True else df[col].transform(lambda x: (x - x.min()) / (x.max() - x.min()))
    return df
```

```python
# normalize numerical columns
total_encoded_tabular = mynorm(total_encoded_tabular, num_list, False)
```

```python
total_encoded_tabular = total_encoded_tabular.drop(columns = (num_list+dates+others))
```
