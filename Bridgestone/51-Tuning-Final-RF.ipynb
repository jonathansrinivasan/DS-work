```python
import numpy as np
import pandas as pd
import sklearn
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
dat15 = pd.read_csv('individuals15.csv')
dat16 = pd.read_csv('individuals16.csv')
dat17 = pd.read_csv('individuals17.csv')
dat18 = pd.read_csv('individuals18.csv')
```

```python
dat15 = dat15.drop('Unnamed: 0',1)
dat16 = dat16.drop('Unnamed: 0',1)
dat17 = dat17.drop('Unnamed: 0',1)
dat18 = dat18.drop('Unnamed: 0',1)
```

```python
model = RandomForestClassifier()
```

### Training: 2015, predicting: 2016

```python
X_train = dat15.drop(columns = ['indiv','label'])
y_train = dat15['label']
X_test = dat16.drop(columns = ['indiv','label'])
y_test = dat16['label']
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict_proba(X_test)
preds = pd.DataFrame(preds)
```

```python
indivs = dat16['indiv']
preds = preds.join(indivs)
pred100k = preds.sort_values(0).head(100000)['indiv'].tolist()
targets = dat16[y_test == 1]['indiv'].tolist()
```

```python
print(f'Model accuracy: {(len(set(pred100k).intersection(set(targets))) / 100000)}')
```

```python
model = RandomForestClassifier(max_depth = 10,n_estimators = 100,min_samples_split = 2,min_samples_leaf = 2)
```

```python
model.fit(X_train, y_train)
```

```python
preds = model.predict_proba(X_test)
preds = pd.DataFrame(preds)
indivs = dat16['indiv']
preds = preds.join(indivs)
pred100k = preds.sort_values(0).head(100000)['indiv'].tolist()
targets = dat16[y_test == 1]['indiv'].tolist()
print(f'Model accuracy: {(len(set(pred100k).intersection(set(targets))) / 100000)}')
```

### Tuning

```python
#max_depth
mds = [None,10,100]
#n_estimators
n_ests = [10,100,150]
#min_samples_split
mss = [2,3,4]
#min_samples_leaf
msl = [1,2,3]
```

```python
X_train = dat15.drop(columns = ['indiv','label'])
y_train = dat15['label']
X_test = dat16.drop(columns = ['indiv','label'])
y_test = dat16['label']
indivs = dat16['indiv']
```

```python
for m in mds:
    for n in n_ests:
        for m2 in mss:
            for m3 in msl:
                model = RandomForestClassifier(max_depth = m,
                                               n_estimators = n,
                                               min_samples_split = m2,
                                               min_samples_leaf = m3)
                model.fit(X_train, y_train)
                preds = model.predict_proba(X_test)
                preds = pd.DataFrame(preds)
                preds = preds.join(indivs)
                pred100k = preds.sort_values(0).head(100000)['indiv'].tolist()
                print(f'Max depth:{m};  N_estimators: {n}; Min_samples_split: {m2}; Min_samples_leaf:{m3}')
                print(f'Model accuracy: {(len(set(pred100k).intersection(set(targets))) / 100000)}')
                print('-----------------------------------------------------------------------------------')
```
