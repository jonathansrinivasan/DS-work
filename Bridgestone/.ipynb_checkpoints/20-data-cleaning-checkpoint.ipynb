```python
df = data.iloc[:, 1:]
```

```python
df.head()
```

### Detecting the unique values for each column

```python
for i in df.columns:
    print(f'({i} : {df[i].nunique()})')
```

```python
top_stores_trans_list = df.STORE_ID.value_counts().nlargest(5).index
```

### Feature : is the store among the top performers regarding total number of transactions in general?

```python
df['store_top_tran'] = np.where(df.STORE_ID.isin(top_stores_trans_list), 1, 0)
```

```python
# sanity check
df['store_top_tran'].value_counts()
```

### Feature : is the store among the top performers regarding total number of TIRES transactions ?

```python
tires = df.query('PROD_GROUP_CODE == 5')
```

```python
top_stores_tires_list = tires.STORE_ID.value_counts().nlargest(5).index
```

```python
top_stores_tires_list
```

```python
df['store_top_tires_tran'] = np.where(df.STORE_ID.isin(top_stores_tires_list), 1, 0)
```

```python
# sanity check

df['store_top_tires_tran'].value_counts()
```
