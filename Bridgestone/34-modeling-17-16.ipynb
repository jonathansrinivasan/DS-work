```python
import pandas as pd
df17=pd.read_csv('individuals17.csv', index_col = 0)
df17.head()
```

```python
df16=pd.read_csv('individuals16.csv', index_col = 0)
df16.head()
```

```python
x_train = df17.drop(columns = ['indiv', 'label'])
y_train = df17['label']
x_test = df16.drop(columns = ['indiv', 'label'])
y_test = df16['label']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
model = RandomForestClassifier()
model.fit(x_train,y_train)
yhat = model.predict(x_test)
```

```python
print("Accuracy Score: ", accuracy_score(y_test, yhat))
```

```python
pred = pd.DataFrame(model.predict_proba(x_test))
pred.head()
```

```python
pred_sort=pred.sort_values(by = [1],axis = 0,ascending = False)
pred_sort.head()
```

```python
total=pd.concat([df16,pred],axis=1)
total_sort=total.sort_values(by = [1],axis = 0,ascending = False)
total_sort.head()
```

```python
import numpy as np
top10000=total_sort.head(10000)
print("The proportion of label 1 within the top 10000 observations with highest probability of label 1 is ",np.mean(top10000["label"]==1))
```
